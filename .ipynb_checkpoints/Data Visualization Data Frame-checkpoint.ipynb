## Imports

```python
import pandas as pd
import nltk
```

## Clean and Format CSV

```python
races = pd.read_csv('Data Schema - Senator Race (5).csv')
```

#### Creating average value for likelihood predictions

```python
## Assigning values to predictions
mapper = {
'Tossup' :3,            
'Likely R' : 1,           
'Lean R' :2,           
'Safe D' :6,          
'Likely D' :5,         
'Safe R' :0,           
'Lean D (flip)': 4,      
'Lean D':4,         
'Likely D (flip)' : 5,   
'Safe D/I' :6   ,     
'Likely R (flip)' :1}
```

```python
races['Cook'] = races['Cook'].map(mapper)
```

```python
races['IE'] = races['IE'].map(mapper)
```

```python
races['Sabato'] = races['Sabato'].map(mapper)
```

```python
races['CNN'] = races['CNN'].map(mapper)
```

```python
races['RCP'] = races['RCP'].map(mapper)
```

```python
races.drop('NYT',1)
```

```python
# Averaging the assigned values to get concensus between prediction sources

races['Predict_avg'] = races[['Cook', 'IE','Sabato','CNN','RCP']].mean(axis=1)
```

```python
# Standardizing them for visualization purposed (1 for Rep, 2 or Und., 3 for Dem)

prediction = []

for race in races['Predict_avg']:
    
    if race < 2:
        race = 1
            
        
    elif race > 4:
        race = 3
            
    else:
        race = 2
        
    prediction.append(race)
        
    
```

```python
races['Predict_ave_st'] = prediction
```

#### Calculating for/against percentage

```python
## Counting the number of bills candidates have vote for and against. 

voteyears = ['1993','1994','1999','2005','2011','2013','2015','2016','2017']
goodvote = 0
goodvote = 0

goodvotes = []
badvotes = []


for candidate in races['Incombant Gun Control Stance (Voted For)']:
    goodvote = 0
    for bill in nltk.wordpunct_tokenize(str(candidate)):
        print(bill)
        if bill in voteyears:
            goodvote += 1
            
    goodvotes.append(goodvote)
            

for candidate in races['Incombant Gun Control Stance (Voted Against)']:
    badvote = 0
    for bill in nltk.wordpunct_tokenize(str(candidate)):
        if bill in voteyears:
            badvote +=1
                
    badvotes.append(badvote)
    
```

```python
races['Vote for Good'] = goodvotes
```

```python
races['Vote for Bad'] = badvotes
```

```python
## Calculating the percentage of for/against votes. 

races['Good vote percentage'] = (races['Vote for Good'] / (races['Vote for Good'] + races['Vote for Bad']) *100).round()
```
