```python
from bs4 import BeautifulSoup
from __future__ import print_function, division
```

```python
# if needed: pip install requests
import requests

url = 'http://www.equibase.com/profiles/Results.cfm?type=Horse&refno=9369383&registry=T&rbt=TB'

response = requests.get(url)
```

```python
response.status_code
```

```python
print(response.text)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# soup.find() returns the first matched tag it finds.
# It searches the entire tree.

# Search for a type of tag by using the tag as a string
# (like 'body','div','p','a') as an argument.

print(soup.find('td'))
```

```python
# Equivalently:
print(soup.td)
```

```python
# Prettier:
print(soup.td.prettify())
```

```python
# soup.find_all() returns a list of all matches

for link in soup.find_all('td'): 
    print(link)
```

```python
# retrieve the url from an anchor tag
soup.find('a')['href']
```

```python
# You can match on an attribute like an id or class.
# Take a look at what the 'mp_box_content' classes
# look like on the webpage, with Inspect Element.

for element in soup.find_all(class_="phone-collapse table-hover sales-history"):
    print(element, '\n')
```

```python
# We can find all the columns in the first mp_box_content table
# by "chaining" `find` and `find_all`.

print(soup.find(class_='phone-collapse table-hover sales-history').find_all('td'))
```

```python
# Domestic Total Gross

## text does an exact match search!
print(soup.find(text="Buyer"))
```

```python
import re 
dtg_string = soup.find(text=re.compile('Buyer'))
print(dtg_string)
```

```python
print(dtg_string.findNextSibling())
```
