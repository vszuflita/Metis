## Scraping Pages

```python
kidshows = {'SesameStreet': 'http://www.azlyricdb.com/artist/Sesame-Street-7560', 
#  FIND THE 'Wiggles': 'http://www.azlyricdb.com/artist/Wiggles-27423', 
# FIND 'YoGabbaGabba': 'https://www.letssingit.com/yo-gabba-gabba-f5cvf/lyrics', 
 'Barney': 'http://www.azlyricdb.com/artist/Barney-1060', 
 'Dora': 'http://www.azlyricdb.com/artist/Dora-The-Explorer-27853', 
'SpongeBob' :'http://www.azlyricdb.com/artist/Spongebob-Squarepants-21635'}
 
```

```python
AllSongs = {}

for key, value in kidshows.items():   
    
    print(key) # Too keep track of progress
    
    soup=BeautifulSoup(requests.get(value).text, "lxml")
    
    for link in soup.findAll('a', attrs={'href': re.compile("/lyrics")}):
        l = (link.get('href')) ### Getting each of the song links from the main page...
        
        soupsong = BeautifulSoup(requests.get('http://www.azlyricdb.com' + l).text, "lxml") ###Requesting each song
        
        header = soupsong.find('h1').getText() ###Getting song title
        print(header) # To keep track of progress 
        
        listy = []
        
        for items in soupsong.findAll('li'): ### Because each lyric is a listed line. 
            items = items.getText()
            if '[' not in items: ###Because we don't want stage direction included
                listy.append(items) ###Adding to a list to compile full song.
            
        AllSongs[header] = listy ### Adding full song to the dictionary 
            
        time.sleep(4) ###So they don't get wise to us! :)
            
    
```
