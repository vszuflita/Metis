## NLP Prep

#### Count Vectorizing (top words)

```python
Songs = SesameSongs_clean['Lyrics'].tolist()
```

```python
Songs
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.feature_extraction.text import CountVectorizer 
```

```python
# CountVectorizer is a class; so `vectorizer` below represents an instance of that object.
vectorizer = CountVectorizer(ngram_range=(1,1), stop_words="english")

# call `fit` to build the vocabulary
vectorizer.fit(Songs)

# then, use `get_feature_names` to return the tokens
print(vectorizer.get_feature_names())

# finally, call `transform` to convert text to a bag of words
x = vectorizer.transform(Songs)
```

```python
x_back = x.toarray()
```

```python
pd.DataFrame(x_back, columns=vectorizer.get_feature_names())
```

```python
vectorizer.get_feature_names()
```

#### Top Bigrams

```python
vectorizerbi = CountVectorizer(ngram_range=(1,2), stop_words="english")

# call `fit` to build the vocabulary
vectorizerbi.fit(Songs)

# finally, call `transform` to convert text to a bag of words
xbi = vectorizerbi.transform(Songs)
```

```python
xbi_back = xbi.toarray()
```
